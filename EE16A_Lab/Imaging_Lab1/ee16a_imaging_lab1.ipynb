### System Overview
Figure 2 is a high-level block model of our imaging system. Here is the system flow: 1) The ambient light sensor converts light energy into electric current; more ambient light leads to more electric current. 2) The current produced by the ambient light sensor then travels through passive electrical components, which translates the current into a detectable signal (a voltage signal). 3) The TI Launchpad reads this signal, and after some linear algebra the voltage signal gets converted into an image. 4) Further improvements are made via software to increase the quality of the image.
<br><br>
<center>
<img src="http://i.imgur.com/RVYM1WH.jpg" style="height:256px" />
<figcaption>Figure 2</figcaption>
</center>

Digital images are composed of a grid of [pixels](http://en.wikipedia.org/wiki/Pixel), each of which represents a single color. To create a digital representation, optical imaging systems measure the energy, equivalent to light in this case, reflected off of objects to determine the values of these pixels. Because different colors reflect different amounts of light, measurements of reflected light intensity can be used to differentiate between colors.
<br/><br/>

<center>
<img src="http://www-inst.eecs.berkeley.edu/~ee16a/sp15/Labs/imaging/nb_img/concept_diagram.png" style="height:256px" />
<figcaption>Figure 3</figcaption>
</center>

Commercial cameras, such as those found in smartphones, use chips called CMOS image sensors to measure millions of pixel values at a time (think of this as millions of devices akin to the ambient light sensor that all fit on an approximately 30mm$^2$ chip). In contrast, our system can only measure one light intensity value at a time, so we will instead use a controlled energy source, the projector, to illuminate specific regions of the image one region at a time. We then use the ambient light sensor to record the energy that reflects off of each region.
<br/><br/>

The ambient light sensor converts light to induce a flow of electrons, and the circuit then outputs a voltage corresponding to light intensity. The specific configuration of the circuit sets the ambient light sensor up so it is most sensitive to changes in light. 
<br/><br/>

This changing voltage is now a set of real numbers and much closer to something we can use to represent our image on a computer. We use a microcontroller (Launchpad) to take samples of the signal and convert it to a set of numbers we can then use in IPython.

<br/>
A voltage is defined as a difference in electric potential between two points. For example, the voltage source we will be using in this lab has a potential difference of $+3.3V$ between the positive and negative terminals of the source. Because a voltage requires two points to define, it is often convenient to define a common point in a circuit as a reference. This reference voltage, commonly referred to as "ground" is defined by convention to be $0V$, and therefore allows us to conveniently describe voltages at any point in the circuit with respect to 0.
